```python
import os
import matplotlib.pyplot as plt
%matplotlib inline
        
from PIL import Image, ImageDraw, ImageFont
import glob
image_list = []

for filename in sorted(glob.glob('C:/Lorraine/OUTPUTS/FINAL_FINAL_OUTPUTS/FACTOR_ANALYSIS/SCREEPLOT/Model1_16/*.png')):    
    im=Image.open(filename)
    draw = ImageDraw.Draw(im)
    # desired size
    #font = ImageFont.truetype('Roboto-Bold.ttf', size=45)
    font = ImageFont.truetype("arial.ttf", size=30)
    # starting position of the message
    (x, y) = (50, 50)
    model = str(filename)
    message = model.split('_', -5)
    message = message[5]
    message = str(message)
    color = 'rgb(0, 0, 0)' # black color
    # draw the message on the background
    draw.text((x, y), message, fill=color, font=font)
    # save the edited image
    im.save(filename)
    image_list.append(im)
      
plt.figure(figsize=(80,140))
plt.margins(0.02)


#plt.title('Model 1: Model 4 Correlation Matrices, Attraction on Left Production on Right\n', verticalalignment='top')
for i, image in enumerate(image_list):
    plt.subplot(8, 4, i+1)
    #plt.subplots_adjust(left=0.1, right=0.2, top=0.2, bottom=0.1)
    plt.imshow(image)
    
plt.savefig('C:/Lorraine/OUTPUTS/FINAL_FINAL_OUTPUTS/FACTOR_ANALYSIS/SCREEPLOT/FactorTotalVariance1-16.png', bbox_inches='tight')
```
